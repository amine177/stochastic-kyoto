# Demonstrating matplotlib

* Plotting 
$$
f :\rm I\!R \to \rm I\!R \\ \quad\quad x \mapsto sin(x)
$$

```python
# inline : shows gaphs integrated in the notebook `inline`
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot') # use grammar of graphics plot style 

x = np.linspace(-3, 3) # create an array of evenly spaced numbers between -3 and 3 inclusive
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

* Plotting
$$ x^k \quad \forall k \in \rm I\!N $$

```python
def powerBuilder(k): # creating a power function builder i.e powerBuilder(2) -> f(x) = x^2 ..
    return lambda x: x**k
```

```python
x - np.linspace(-2, 2)

plt.plot(x, powerBuilder(1)(x), label='$f(x)=x$')     # f(x) = x
plt.plot(x, powerBuilder(2)(x), label='$f(x)=x^2$')   # f(x) = x^2
plt.plot(x, powerBuilder(3)(x), label='$f(x)=x^3$')   # f(x) = x^3

plt.legend()       # show legend
plt.xscale('log')  # logscale the x axis
plt.yscale('log')  # logscale the y axis
plt.xlabel('x')
plt.ylabel('f(x)')

plt.show()
```

* Plotting a histogram
$$ P(X) \in [0, 1] \quad \forall X \sim U(0, 1) $$

```python
N = 10**5
np.random.seed(0)  # initialized the prng with a seed of 0
R = np.random.rand(N) # create N random values in [0, 1]
plt.hist(R, bins=100, normed=True) # create the histogram plot which have 100 bins which are normalized
plt.show()
```

* Plotting 10 random walks with 10000 steps
$$
\{X_t\}_{t \in \rm I\!N}, X_t = \begin{cases}
1 & \text{with probability 0.5} \\
-1 & otherwise
\end{cases}
$$

```python
NSTEP = 10**4
plt.xlabel('t')
plt.ylabel('X_t')
for nseed in range(10):
    np.random.seed(nseed)   # feed seed to get a new trajectory
    X_t = np.random.choice([-1, 1], NSTEP) # generate 10000 steps
    trajectory = np.cumsum(X_t)            # do a cumulative sum of X_t for t in [0, 10**4 - 1]
    plt.plot(trajectory)
```
